## 神经网络

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 读取数据集
data = pd.read_pickle('../以天粒度预测/data4_5.pkl')

# 计算每列数据的众数
mode_values = data.mode()

# 使用众数填补缺失值
data = data.fillna(mode_values.iloc[0])

# 将订单需求量为0的数据删除
d = data[data['订单需求量']!=0]
data = d

# 将输入变量和输出变量分开
X = data.iloc[:, :5].values  # 假设前五列是输入变量
X = np.concatenate((X, data.iloc[:, 6:].values), axis=1)  # 将第七列及其后面的列作为输入变量添加到X中
y = data.iloc[:, 5].values  # 第六列是输出变量

# 将数据集分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)

# 将输入变量进行标准化处理
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# 构建神经网络模型
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
model.add(tf.keras.layers.Dense(units=32, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))

# 编译模型
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# 绘制损失变化曲线
plt.figure(figsize=(18, 12))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 使用训练好的模型进行预测
y_pred = model.predict(X_test)

# 计算RMSE和MAPE值
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = mean_absolute_percentage_error(y_test, y_pred) * 100

print("RMSE值为：", rmse)
print("MAPE值为：", mape)

```

## LSTM

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 读取数据集
data = pd.read_pickle('../以天粒度预测/data4_5.pkl')

# 计算每列数据的众数
mode_values = data.mode()

# 使用众数填补缺失值
data = data.fillna(mode_values.iloc[0])

# 将订单需求量为0的数据删除
d = data[data['订单需求量']!=0]
data = d

# 将输入变量和输出变量分开
X = data.iloc[:, :5].values  # 假设前五列是输入变量
X = np.concatenate((X, data.iloc[:, 6:].values), axis=1)  # 将第七列及其后面的列作为输入变量添加到X中
y = data.iloc[:, 5].values  # 第六列是输出变量

# 将数据集分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
y_train = y_train.astype(np.float64)
y_test = y_test.astype(np.float64)

# 将输入变量进行标准化处理
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# 将训练集和测试集转化为LSTM的输入格式
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# 构建LSTM模型
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.LSTM(units=64, activation='relu', input_shape=(1, X_train.shape[2])))
model.add(tf.keras.layers.Dense(units=32, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))

# 编译模型
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))

# 绘制损失变化曲线
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 使用训练好的模型进行预测
y_pred = model.predict(X_test)

# 计算RMSE和MAPE值
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = mean_absolute_percentage_error(y_test, y_pred) * 100

print("RMSE值为：", rmse)
print("MAPE值为：", mape)
```

```python
# 读取数据集
data = pd.read_pickle('../以天粒度预测/data4_5.pkl')

# 计算每列数据的众数
mode_values = data.mode()

# 使用众数填补缺失值
data = data.fillna(mode_values.iloc[0])

# 将订单需求量为0的数据删除
d = data[data['订单需求量']!=0]
data = d

# 将输入变量和输出变量分开
X = data.iloc[:, :5].values  # 假设前五列是输入变量
X = np.concatenate((X, data.iloc[:, 6:].values), axis=1)  # 将第七列及其后面的列作为输入变量添加到X中
y = data.iloc[:, 5].values  # 第六列是输出变量

# 将数据集分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
y_train = y_train.astype(np.float64)
y_test = y_test.astype(np.float64)

import lightgbm as lgb

# 创建LightGBM模型
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'n_estimators':200,
    'learning_rate':1,
    'subsample':0.8,
    'colsample_bytree':0.7,
    'max_depth':15,
    'num_leaves':50,
    'min_child_weight':200

}
model = lgb.LGBMRegressor(**params)

# 在训练集上拟合模型
model.fit(X_train, y_train)

# 在测试集上进行预测
y_pred = model.predict(X_test)

# 计算预测结果的均方误差和MAPE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('均方误差：', rmse)

from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
print('MAPE：',mape)
```

```python
import matplotlib.pyplot as plt

# 计算真实值与预测值之间的误差
errors = y_pred-y_test

# 绘制误差曲线图
fig, ax = plt.subplots(figsize=(18, 8))
ax.plot(errors[:500], color='#1f77b4', linewidth=1.5, label='误差')
ax.axhline(y=0, color='#d62728', linestyle='--', linewidth=1, label='基准线')
ax.set_xlabel('样本序号', fontsize=25)
ax.set_ylabel('误差', fontsize=25)
# 设置坐标轴刻度的字体大小
plt.xticks(fontsize=21)
plt.yticks(fontsize=21)
ax.set_title('XGBoost模型真实值与预测值误差曲线图', fontsize=28)
ax.legend(fontsize=23)
plt.savefig('xgb误差.jpg',dpi=200)
plt.show()
```
